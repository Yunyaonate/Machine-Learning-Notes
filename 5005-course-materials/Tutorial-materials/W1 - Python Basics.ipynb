The if statement is controlling the program by changing its behaviour depending on the user's input. This why it is called a control structure.

if statements (and other control structures) control a chunk of code called a block. In Python, a statement is indented to indicate it belongs to a block. For example:

```python
food = input("What food do you like? ")
if food == "cake":
 print("Wow, I love cake too!")
 print("Did I tell you I like cake?")
```

The block is often called the body of the control structure. Every control structure in Python has a colon followed by an indented block.

You will notice in our examples that we are using two equals signs to check whether the variable is equal to a particular value:

```python
name = "Grok"
if name == "Grok":
  print("Hello, friend")
```

A double == is used for comparison. This is what we do to check whether two things are equal. The second line of the program above is checking whether the variable name is equal to "Grok" using a double equals sign.

if statements allow you to make yes or no decisions. In Python (and most programming languages) these are called True and False.

```python
name = "Grok"
if name == "Grok":
  print("Hello, don't I know you?")
```

If we change name the expression will evaluate to False and the block of code will not be run:

All of the examples so far have done something when a condition is True, but nothing when the condition is False. In the real world we often want to do one thing when a condition is True and do something different when a condition is False.

For example, we may want to welcome a friend, but find out more about someone we don't know. We can do that using an else clause:

```python
name = input("Who goes there? ")
if name == "Frodo":
  print("Welcome, we're expecting you.")
else:
  print("State your business, stranger.")
```

If the user enters "Frodo" the first message will be printed. In all other cases, the second message will be printed.

Using the else clause can make code neater and easier to read. For example, let's go back to our raining example from the first slide:

```python
raining = input("Is it raining (yes/no)? ")
if raining == "yes":
  print("You should take the bus to work.")
if raining == "no":
  print("You should walk to work.")
```

We can simplify this using else:



```python
raining = input("Is it raining (yes/no)? ")
if raining == "yes":
  print("You should take the bus to work.")
else:
  print("You should walk to work.")
```

Here, either the first or second print statement will be executed but not both. Notice that the else clause must be followed by a : character, just like the if statement.

As well as being neater, the else statement is more efficient because it does not have to perform another conditional evaluation.

The if/else statement is good for making decisions when there are only two options (friend or foe, true or false, wet or dry). When there are more than two possibilities, the results can be confusing. This is something to watch out for.

# looping

Computers are not very smart, but they are extremely fast  at doing exactly what you tell them to do over, and over, and over again.

Until now, the only way you can repeatedly run code is to duplicate it (almost by copy + paste). To print each character and its position in a string:

```python
name = input('Enter your name? ')
print(name[0])
print(name[1])
print(name[2])
print(name[3])

# but what if the name entered is not 4 characters long?
# The program either stops with an IndexError (on short names) or doesn't print out all of the characters (on long names). 
# How can you solve the problem for names of any length? The answer is loops!
```

Let's be reckless  we're going to use a loop and then explain it. We can get the Python interpreter to repeat things using loops:

```python
name = input('Enter your name? ')
for c in name:
  print(c)
```

It's as easy as that, we're off and looping!

The for loop steps through a known list of elements, in this case the list of characters in the name entered by the user. If there are 3 letters in the name, then the for loop will run 3 times, one iteration for each character. For each pass through the loop the variable named in the for statement will be assigned the value of the current item in the list. So for the loop above (if the user enters 'Bob'), the value assigned to c will change from 'B' to 'o' to 'b'.

The for loop sets up a variable, in this case we've called it c, which is short for character. We can call the variable any name, perhaps character, though as a convention these names are often quite short, and for numbers are often called i (for integer) or n (for number).

```python
name = input('Enter your name? ')
for character in name:
  print(character)
```


We may want successive print() calls to continue output on the same line as the last print() call. We can accomplish this by specifying an end argument that you pass to the print() function. This will insert a space and cause the next print() call to continue this line and not start another. For example:

```python
for i in "Hello":
  print(i, end=" ")
print('Next line')
```

Notice how the Next line output isn't printed on the next line at all, but at the end of that line. We can do better by building up a new string to print out once the loop has finished.

We start with an empty string: output, and in each iteration of the loop we add on an extra letter:

```python
output = ''
for i in "Hello":
  output += i
  print(output)
print(output)
print('Next line')
```

Often we know how many times we want to repeat things, and we want a loop that runs a particular number of times. We can do this using the range() function. Let's give it a go:

```python
for i in range(5):
  print(i)
```

A really important note about range is that it starts from zero and going up to n - 1 (rather than going from 1 to n like you normally would).

The for loop steps through a known list of elements, in this case the list of numbers generated by the range() function. For each pass through the loop the variable named in the for statement will be assigned the value of the current item in the list. So in the example above, the value assigned to i changes from 0 to 1 to 2.

With a single argument the first number in the range() is always 0. We can change that by adding an optional second argument when we call range().

```python
for n in range(5, 8):
  print(n)
```


With two arguments range() returns a list starting at the first value and going up to, but not including, the last value. If we wanted to include 8 in the loop we would have to do this:

```python
for n in range(5, 9):
  print(n)
```

We can also pass in a third argument to range, the 'step', which lets us step through the numbers by an amount other than 1. For example, to print even numbers only:

```python
for i in range(2, 12, 2):
  print(i)
```


In English this means "Print the numbers from 2 up to (but not including) 12, in steps of 2".

You can also count using negative numbers:

```python
for i in range(-10, -7, 1):
  print(i)
```

You can count backwards by passing a negative step value into the range function:



```python
for i in range(5, 0, -1):
  print(i)	
```

This loop counts backwards with i being assigned each of [5,4,3,2,1] in turn and then used as an argument to the print statement.

range() will ignore input when it does not make sense. For instance, you can't "automatically" count backwards; your start and end numbers need to make sense:

```python
for i in range(12, 3):
  print(i)

# there is nothing to print out!
```


It's worth taking a moment to combine what we have learnt here, how to print out a series of numbers, with what we learnt earlier, how to print output on one line. One way to do this is to build up a string, bit by bit, to print out. First, we'll need to set up a variable outside the loop that we can add to from inside the loop:

```python
output = ""
for i in range(10):
  output = output + str(i) + ' '
print(output)

# although it is invisable in colab, but it has an extra space in the end
# you also check the length of the string
```

Notice that this leaves an extra space on the end of this string. We can get rid of this extra space by using rstrip which strips whitespace from the right of the string, or by using string slicing. The slice output[:-1] includes everything up to but not including the last character.

```python
output = ""
for i in range(10):
  output = output + str(i) + ' '
print(output[:-1])
print(output.rstrip())
```

We can use for loops to do maths. Let's calculate the sum of all integers less than 100:



```python
result = 0
for i in range(100):
  result = result + i
print(result)
```

The first line of this program, result = 0, sets the initial value of result to 0. Look what happens if we leave it out:

```python

for i in range(100):
  result = result + 1
print(result)
```

Inside the loop body the line result = result + 1 cannot execute if the variable result is not already defined. By assigning it to 0 before we start the loop we let Python know that it is a variable and also that it will be used for integers.

We can use for and range() to access characters of a string inside a loop. For example, we can print out every character in a string, one character per line, and also print out the index of each letter like this:

```python
y = 'hello'
for i in range(len(y)):
  line = "Letter " + str(i) + " is " + y[i]
  print(line)
```

A shorter part of the string (called a substring) can be accessed by using an extended square bracket notation with two indices called a slice:

```python
x = "hello world"
print(x[0:5])
print(len(x))
print(x[6:11])
```


A slice starts from the first index and includes characters up to but not including the second index. So, the slice x[0:5] starts from the x[0] and goes up to x[5] (the space) but doesn't include it. There are a couple of shortcuts too:

```python
x = "hello world"
print(x[:5])
print(x[6:])
```

When the first index is zero (when taking a prefix of the sequence) the zero can be left out (like x[:5]). When the last index is the length of the string, (when taking a suffix of the sequence) the last index can be left out (like x[6:]).

Negative indices work in the same way that they do for indexing. Slicing is slightly more general than indexing. The index values must fall within the length of the string, but slice indices do not have to. If the slice indices fall outside the length of the string then the empty string will be returned for all or part of the slice.

Strings are also immutable, which means that once the string is constructed, it cannot be modified. For example, you cannot change a character in an existing string by assigning a new character to a subscript:

```python
x = 'hello world'
x[0] = 'X'
```

You must instead create a new string by slicing up the string and concatenating the pieces, and then assigning the result back to the original variable:

```python
x = 'hello world'
x = 'X' + x[1:]
print(x)
```

# Data type and list

We have already done some basic maths using Python, but everything we have done so far used whole numbers, or integers. If we want to use fractions, or numbers with a decimal place, we need floating point numbers, or floats.

Like strings and integers, floats are a separate data type and they have different behaviour (even though you and I just think of them all as numbers).

You may have even used floats before without realising it if you tried out division. In Python 3, normal division will return a float:
 


```python
print(10/3)
```

Even in cases where there is no decimal point required, the number that is returned and printed out is still a float:

```python
print(10/2)
```

Changing one type of data to another is called casting. You can cast an integer to a floating point number using the float() function:

```python
i = 10
print(i)
f = float(i)
print(f)
```

When we print out the float f there is a decimal place, and when we print out the integer i there isn't one.

You can convert a string containing a number directly to a float:

```python
s = '10.5'
f = float(s)
print(f+2)
print(s+2)
```

The difference between the string s = '10.5' and the float f = float(s) is that trying to do maths with the string will cause an error. Try running the program above to see this.

You can change a float to an integer using the int function:



```python
f = 10/2
i = int(f)
print(i)
```

You've probably noticed that this looks just like converting strings to integers using the int function:



```python
s = '5'
i = int(s)
print(i)
```

One important thing to note is that if you cast a floating point number to an integer, it will always round down to the integer below. So for example:



```python
print(int(4.8))
print(int(2.2))
```

Here 4.8 becomes 4 and 2.2 becomes 2. This is probably a bit surprising, because it is different to the rounding you do in maths, where 4.8 would be rounded to 5.

```python
remainder = 10%3
print(remainder)
```

#More on looping

So far we have looked at how to repeat things when we know how many times we want to repeat something. In this module we will look at repeating things while a particular condition is true. Not surprisingly, the way of doing this in Python is with while loops.

Let's start with a simple example:

```python
i = 0
while i < 3:
  print(i)
  i = i + 1
```

In English, this loop says "While i is less than 3, print i.".

Before the loop starts, i is initialised to zero. Zero is less than three, so the loop runs the first time. The final line of the loop increments i by adding one to it. Each time the loop runs, it checks whether the condition i < 3 is still True. If it is, it runs another iteration.

There are four things to think about when writing your own loops:

Starting: Are there any variables we need to setup before we start a loop? This is often called loop initialisation. In our example, this was the line:

```python
i = 0
```

Stopping: How do we know when to stop? This is called the stopping or termination condition. In our example, this was the line:

```python
while i < 3:
```

Doing: What are the instructions we want to repeat. This will become the body of the loop control structure. In our example, this is:

```python
  print(i)
```

Changing: Do the instructions change in any way between each repetition. For example, we might need to keep track of how many times we have repeated the body of the loop, that is, the number of iterations. In our example, this is:

```python
  i = i + 1
```

You can use these four as a checklist to make sure the loops you write are doing what you want them to.

Just like in for loops we can take steps that are greater than 1. For example, the count in odd numbers only:

```python
i = 1
while i < 7:
  print(i)
  i = i + 2
```

The initialisation involves setting the loop variable to one with i = 1. The loop will terminate when i is no longer smaller than seven (i < 7).

There are two statements in the body. The first body statement is print, the second is i = i + 2 which increments our loop variable by 2. The result is the loop will now print out all odd numbers less than seven.

We can also count backwards using while loops. For example:



```python
j = 6
while j > 3:
  print(j)
  j = j - 1
```

Notice this time around the loop starts at a larger value and counts down, because we're using j = j - 1 rather than j + 1. The termination condition is now greater than 3, rather than a less than 3.

A while loop is very flexible. How you choose to initialise, run and terminate the loop is up to you. You can use whatever variables, in whatever combination and with whatever values you like.

We can access items in a dictionary using square brackets, like lists and strings:



```python
sounds = {'dog': 'a barker', 'cat': 'a meower', 'bird': 'a tweeter'}
print(sounds['dog'])
```

Here, `sounds['dog']` returns the corresponding value 'a barker' since 'dog' is in the dictionary.

What happens if the key is not in the dictionary?

```python
sounds = {'dog': 'a barker', 'cat': 'a meower', 'bird': 'a tweeter'}
print(sounds['mouse'])
```

To tell if a key is in a dictionary we can use the in operator, just like with strings and lists:

```python
sounds = {'dog': 'a barker', 'cat': 'a meower', 'bird': 'a tweeter'}
if 'cat' in sounds:
  print('cat is in the dictionary')
if 'mouse' in sounds:
  print('mouse is in the dictionary')
```

```python
sounds = {'dog': 'a barker', 'cat': 'a meower', 'bird': 'a tweeter'}
for animal in sounds:
  print('a', animal, 'is', sounds[animal])
```

You can add and modify items in a dictionary by assigning a new key-value pair, either over-writing an existing one, or adding a new definition:

 


```python
sounds = {'dog': 'a barker', 'cat': 'a meower', 'bird': 'a tweeter'}
animal = 'cat'
print('A', animal, 'is', sounds[animal])

# Change the cat definition:
sounds['cat'] = 'a purrer'
print('A', animal, 'is now', sounds[animal])

# Add a new definition for mouse:
sounds['mouse'] = 'a squeaker'
animal = 'mouse'
print('A', animal, 'is', sounds[animal])

print('Final dictionary contents:')
print(sounds)
```

One very common way of creating a dictionary is looping through a file or input from the user. We start with an empty dictionary and add key-value pairs. Here's an example code snippet:

```python
# First, we'll set up an empty dictionary:
phonebook = {}
# and read in our first line:
line = input('Name and number: ')

#Now we'll use a while loop to continue to read input from the user until a blank line is entered.
while line:
  #In this while loop, we will split up the line into name and number using line.split():
  parts = line.split()
  name = parts[0]
  number = parts[1]
  # and add it to the dictionary called phonebook
  phonebook[name] = int(number)
  # before reading in another line of input, and continuing the while loop:
  line = input('Name and number: ')

#Then finally, printing out the phonebook:
print(phonebook)
```

# more data structures (tuple and set)

**Tuple**:

Tuples are a data structure in Python which are more or less a read-only list. That is, like strings, tuples are immutable. All of the standard read-only operations that you can perform on a list can be performed on a tuple, including finding its length, slicing, indexing, and unpacking.

Tuples are notated just like lists, except that the items of the tuple are enclosed in parenthesis instead of square brackets.

```python
places = {'buried treasure': (2, 5), 'frisbee': (1, 1)}
print(places['buried treasure'])
places['frisbee'] = (3, 3)
print(places)
```

You can convert any sequence type to a tuple using the tuple function. This can also be used without any arguments to construct an empty tuple.

 


```python
times1 = [43, 45, 49, 52, 55]
times2 = tuple(times1)
times1.append(60)
print(type(times1), times1)
print(type(times2), times2)
```

There is a gotcha to be aware of when creating tuples containing a single element. Python interprets an object surrounded by a single set of parentheses as grouping parentheses and just discards them. To create a single element tuple, a trailing comma is needed. The following example demonstrates this:

```python
x = ()         # empty tuple
print(type(x), x)
x = (9, 4, 1)  # 3 element tuple
print(type(x), x)
x = (9,)       # 1 element tuple
print(type(x), x)
x = (9)        # the integer 9
print(type(x), x)
```

Tuples are **immutable**, an exception is raised if you try to perform a modification operation upon them.

```python
times = (43, 45, 49, 52, 55)
times[2] = 10
```

If you want to modify an element of a tuple, you will need to construct a new tuple and assign the result back to the same variable.

 


```python
times = (43, 45, 49, 52, 55)
times = times[:2] + (10,) + times[3:]
print(times)
```

**Comparing tuples**:

Like most other builtin Python data types, tuples can be compared using the standard comparison operators (<, >, etc). Tuples are compared element-wise. This means that the first item from both tuples are compared. If their values are not equal, then the appropriate tuple is returned. If they are equal, then the second element from both tuples is compared, and this process is repeated.

```python
a = (5, 'hello')
b = (5, 'world')
c = (4, 'hello')
print(a > b, a > c, b > c)

items = [a, b, c]
items.sort()
print(items)
```

A consequence of this is that tuples can only be compared if their pairwise items have the same types. Attempting to compare two objects of different types raises a `TypeError`:

```python
a = (5, 'hello')
b = (5, 3.14)
print(a > b)
```

**Set**:

The set data structure in Python models the mathematical definition of a set: an unordered collection of *unique* objects. The easiest way to think about sets is that they are like dictionaries without values; that is, the keys in a dictionary are a set. Mathematical set operations:

Mathematical sets have a number of standard operations that can be performed on two sets. Python sets have support for the union, intersection, and difference operations.

The union can be thought of as "adding" the two sets together. It returns a set with all of the elements that are in either set. The intersection of two sets returns a set of all the elements which are in both sets. The difference of two sets can be thought of as "subtracting" one set from the other. It returns the elements in the first set minus the elements in the second set. Let's have a look at an example:

```python
a = {1, 2, 3}
b = {3, 4, 5}
print(a | b)  # Union
print(a & b)  # Intersection
print(a - b)  # Difference
print(b - a)  # Difference
print(a ^ b)  # Symmetric difference
```

# Matrix Multiplication

For deep learning, matrix calculation is an important trick to speed up the neural network training progresses (which we will learn late :) ). For now, we need to learn how to use matrix/vector and calculate them in python.

```python
import random

#define a matrix with dimension M * N
def generate_matrix(m, n):
    ls = []
    for i in range(m): #generate outer dimension
        sub_list = []
        for j in range(n): #generate inner dimension
            sub_list.append(random.randint(10, 20)) #random int in [10 to 20]
        ls.append(sub_list)
    return ls

generate_matrix(5, 5)
```

if we two matrix:

**A** =

1,2,3

4,5,6

and **B** =

7, 8

9, 10

10, 12

A is a 2x3 matrix and B is a 3x2 matrix then their multiplication result is 2x2.

The result is:

58, 64

139 154

how to get the answer?

58 (1, 2, 3)  (7, 9, 11) = 17 + 29 + 311 = 58 note: each result is calculated by dot product

64: (1, 2, 3)  (8, 10, 12) = 18 + 210 + 312 = 64

139: (4, 5, 6)  (7, 9, 11) = 47 + 59 + 611 = 139

153: (4, 5, 6)  (8, 10, 12) = 48 + 510 + 612 = 154
notes: given any two matrices m1 and m2, the size of m1 is H1 x W1, and the 
size of m2 is H2 x W2. For matrix multiplication of m1 and m2, e.g., m1 x m2, we need to make sure W1 == H2, and the size of the result will be H1 x W2. Like m2 x m1, we need to make sure W2 == H1, and the result will be H2 x W1.

As we see, m1 x m2 != m2 x m1.

```python
# matrix [i,j] * [j,k]

def matrix_multiply(m1, m2):
    if len(m1[0]) != len(m2): #dimension illegal check
        return None
    result = []
    for i in range(len(m1)):
        sub = []
        for j in range(len(m2[0])):
            v = 0
            for k in range(len(m2)):
                v += m1[i][k] * m2[k][j]
            sub.append(v)
        result.append(sub)
    return result

matrix1 = generate_matrix(3, 4)
matrix2 = generate_matrix(4, 5)

print("the matrix 1 is:")
print(matrix1)
print()
print("the matrix 2 is:")
print(matrix2)
print()
print("the result of matrix multiplication")
matrix_multiply(matrix1, matrix2)
```
